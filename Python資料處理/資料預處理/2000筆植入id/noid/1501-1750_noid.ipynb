```python
import pandas as pd
import numpy as np

table1=pd.read_csv('/Users/humphreyshen/Desktop/2021 Database Project/003數據預處理/2000筆活動資料植入ID工作/原始csv檔區/1501-1750(原始).csv')
table1
```

```python
table2=pd.read_csv('/Users/humphreyshen/Desktop/2021 Database Project/001總表下載數據/df_orginal_withoutblank.csv')
table2.head()
```

```python
table2_clean=table2.replace({'\r':''},regex=True)
table2_clean
```
