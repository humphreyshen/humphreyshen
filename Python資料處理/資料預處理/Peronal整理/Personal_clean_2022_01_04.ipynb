```python
import pandas as pd
import numpy as np
```

```python
Personal_pre=pd.read_csv('/Users/humphreyshen/Desktop/2021 Database Project/001總表最新資料/2021-12-29-updated_csv/dDt_alumnidb_personal .csv')
Personal_pre.head()
```

```python
Personal_pre.count()
```

```python
Personal_birthday=Personal_pre.dropna(subset=['birthday'])
Personal_birthday
```

```python
Personal_cell_drop=Personal_birthday.dropna(subset=['cell'])
Personal_cell_drop
```

```python
Personal_cell_drop.count()
```

```python
Personal_cell_dup_drop=Personal_cell_drop.drop_duplicates(subset=['cell']).replace({'\r':''},regex=True)
Personal_cell_dup_drop
```

```python
Personal_cell_dup_drop.count()
```

```python
Personal_cell_dup_drop.to_csv('2022-01-04_birthday.csv',encoding='utf-8-sig')
```

```python
Personal_birthday_email=Personal_birthday.dropna(subset=['email'])
Personal_birthday_email.count()
```

```python
Personal_birthday_email_duplicated=Personal_birthday_email.drop_duplicates(subset=['email']).replace({'\r':''},regex=True)
Personal_birthday_email_duplicated.count()
```

```python
Personal_birthday_email_duplicated.to_csv('2022-01-04_birthday_email.csv',encoding='utf-8-sig')
```
