```python
import pandas as pd
import numpy as np

liaison_pre=pd.read_csv('/Users/humphreyshen/Desktop/2021 Database Project/班級聯絡人/班級聯絡人-總表.csv')
liaison_preselect=liaison_pre.iloc[:,0:11]
liaison_select=liaison_preselect.dropna(subset=['校友編號'])
liaison_select.head()
```

```python
liaison_select_rename=liaison_select.rename({'序號':'id','校友編號':'alumni_id','畢業年':'grad_year','科系':'dept','班級':'class','姓名':'zh_name','聯絡電話':'cell','信箱':'email','可觸及人數':'contact','生日':'birthday'}, axis=1)
liaison_select_rename.head()
```

```python
liaison_select_rename.head(30)
```

```python
liaison_clean=liaison_select_rename.replace('\n','',regex=True)
liaison_clean.head(30)
```

```python
liaison_replace=liaison_clean.replace('   ','0',regex=True)
liaison_replace[['contact']]=liaison_replace[['contact']].fillna(value=0)
liaison_replace.head()
```

```python
convert_dict={'grad_year':int,'liaison':int}
liaison_replace=liaison_replace.astype(convert_dict)
liaison_replace.head()
```

```python
liaison_replace[['grad_year']]=liaison_replace[['grad_year']]+1911
liaison_replace.head()
```
